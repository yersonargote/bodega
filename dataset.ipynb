```python
# Extracting the year, month and day from the 'FECHA SUSCRIPCION', 'FECHA ACTA DE INICIO' and 'FECHA CREACIÓN' columns
df['FECHA SUSCRIPCION'] = pd.to_datetime(df['FECHA SUSCRIPCION'])
df['FECHA SUSCRIPCION_anio'] = df['FECHA SUSCRIPCION'].dt.year
df['FECHA SUSCRIPCION_mes'] = df['FECHA SUSCRIPCION'].dt.month
df['FECHA SUSCRIPCION_dia'] = df['FECHA SUSCRIPCION'].dt.day

df['FECHA ACTA DE INICIO'] = pd.to_datetime(df['FECHA ACTA DE INICIO'])
df['FECHA ACTA DE INICIO_anio'] = df['FECHA ACTA DE INICIO'].dt.year
df['FECHA ACTA DE INICIO_mes'] = df['FECHA ACTA DE INICIO'].dt.month
df['FECHA ACTA DE INICIO_dia'] = df['FECHA ACTA DE INICIO'].dt.day

df['FECHA CREACIÓN'] = pd.to_datetime(df['FECHA CREACIÓN'])
df['FECHA CREACIÓN_anio'] = df['FECHA CREACIÓN'].dt.year
df['FECHA CREACIÓN_mes'] = df['FECHA CREACIÓN'].dt.month
df['FECHA CREACIÓN_dia'] = df['FECHA CREACIÓN'].dt.day
```

```python
# Create a dictionary to map the names of months with their respective number in Spanish, the key is the number and the value is the name of the month
months = {1:'Enero', 2:'Febrero', 3:'Marzo', 4:'Abril', 5:'Mayo', 6:'Junio', 7:'Julio', 8:'Agosto', 9:'Septiembre', 10:'Octubre', 11:'Noviembre', 12:'Diciembre'}

# Create a new column name 'FECHA SUSCRIPCION_mes_nombre' and map the names of months with their respective number using the dictionary created above
df['FECHA SUSCRIPCION_mes_nombre'] = df['FECHA SUSCRIPCION_mes'].map(months)

# Create a new column name 'FECHA ACTA DE INICIO_mes_nombre' and map the names of months with their respective number using the dictionary created above
df['FECHA ACTA DE INICIO_mes_nombre'] = df['FECHA ACTA DE INICIO_mes'].map(months)

# Create a new column name 'FECHA CREACIÓN_mes_nombre' and map the names of months with their respective number using the dictionary created above
df['FECHA CREACIÓN_mes_nombre'] = df['FECHA CREACIÓN_mes'].map(months)
```

```python
# Create a dictionary to map the months with their respective trimester number, the key is the name of the month and the value is the trimester number
trimesters = {'Enero':1, 'Febrero':1, 'Marzo':1, 'Abril':2, 'Mayo':2, 'Junio':2, 'Julio':3, 'Agosto':3, 'Septiembre':3, 'Octubre':4, 'Noviembre':4, 'Diciembre':4}

# Create a new column name 'FECHA SUSCRIPCION_trimestre' and map the months with their respective trimester number using the dictionary created above
df['FECHA SUSCRIPCION_trimestre'] = df['FECHA SUSCRIPCION_mes_nombre'].map(trimesters)

# Create a new column name 'FECHA ACTA DE INICIO_trimestre' and map the months with their respective trimester number using the dictionary created above
df['FECHA ACTA DE INICIO_trimestre'] = df['FECHA ACTA DE INICIO_mes_nombre'].map(trimesters)

# Create a new column name 'FECHA CREACIÓN_trimestre' and map the months with their respective trimester number using the dictionary created above
df['FECHA CREACIÓN_trimestre'] = df['FECHA CREACIÓN_mes_nombre'].map(trimesters)

# Create a dictionary to map the months with their respective semester number, the key is the name of the month and the value is the semester number
semesters = {'Enero':1, 'Febrero':1, 'Marzo':1, 'Abril':1, 'Mayo':1, 'Junio':1, 'Julio':2, 'Agosto':2, 'Septiembre':2, 'Octubre':2, 'Noviembre':2, 'Diciembre':2}

# Create a new column name 'FECHA SUSCRIPCION_semestre' and map the months with their respective semester number using the dictionary created above
df['FECHA SUSCRIPCION_semestre'] = df['FECHA SUSCRIPCION_mes_nombre'].map(semesters)

# Create a new column name 'FECHA ACTA DE INICIO_semestre' and map the months with their respective semester number using the dictionary created above
df['FECHA ACTA DE INICIO_semestre'] = df['FECHA ACTA DE INICIO_mes_nombre'].map(semesters)

# Create a new column name 'FECHA CREACIÓN_semestre' and map the months with their respective semester number using the dictionary created above
df['FECHA CREACIÓN_semestre'] = df['FECHA CREACIÓN_mes_nombre'].map(semesters)
```

```python
# From date to day of the week
df['FECHA SUSCRIPCION_dia_semana'] = df['FECHA SUSCRIPCION'].dt.day_name()

df['FECHA ACTA DE INICIO_dia_semana'] = df['FECHA ACTA DE INICIO'].dt.day_name()

df['FECHA CREACIÓN_dia_semana'] = df['FECHA CREACIÓN'].dt.day_name()

# Create a dictionary to map the days of the week with their respective Spanish name, the key is the name of the day and the value is the Spanish name
days = {'Monday':'Lunes', 'Tuesday':'Martes', 'Wednesday':'Miércoles', 'Thursday':'Jueves', 'Friday':'Viernes', 'Saturday':'Sábado', 'Sunday':'Domingo'}

# Replace the names of the days of the week with their respective Spanish name using the dictionary created above
df['FECHA SUSCRIPCION_dia_semana'] = df['FECHA SUSCRIPCION_dia_semana'].replace(days)

df['FECHA ACTA DE INICIO_dia_semana'] = df['FECHA ACTA DE INICIO_dia_semana'].replace(days)

df['FECHA CREACIÓN_dia_semana'] = df['FECHA CREACIÓN_dia_semana'].replace(days)
```

```python
# Create a dictionary to map the values of the 'VALOR INIC.CONTRATO' column, from 0 to 10000000 is range 'bajo', from 10000000 to 100000000 is range 'medio' and from 100000000 to 674989151716 is range 'alto'
values = {'bajo': range(0, 10000000), 'medio': range(10000000, 100000000), 'alto': range(100000000, 674989151716)}

# Create a new column named 'RANGO VALOR' that if the 'VALOR INIC.CONTRATO' column is in the range 'bajo', 'medio' or 'alto' then it will be assigned the respective value
df['RANGO VALOR'] = df['VALOR INIC.CONTRATO'].apply(lambda x: next((k for k, v in values.items() if x in v), 0))
```
